```python
import ipywidgets as widgets
from IPython.display import display, clear_output

class TurnoApp:
    def __init__(self):
        self.turnos = []
        self.boton_reiniciar_visible = False

        # Crear widgets
        self.input_turno = widgets.Text(description="Nuevo turno:")
        self.boton_agregar = widgets.Button(description="Agregar turno", style={'description_width': 'initial'})
        self.boton_limpiar = widgets.Button(description="Limpiar lista", style={'description_width': 'initial'})
        self.boton_reiniciar = widgets.Button(description="Reiniciar", style={'description_width': 'initial'})
        self.lista_turnos_output = widgets.Output()

        # Configurar eventos
        self.boton_agregar.on_click(self.agregar_turno)
        self.boton_limpiar.on_click(self.mostrar_confirmacion_limpiar)
        self.boton_reiniciar.on_click(self.reiniciar_app)

        # Mostrar widgets
        self.mostrar_widgets()

    def mostrar_widgets(self):
        display(self.input_turno)
        display(widgets.HBox([self.boton_agregar, self.boton_limpiar, self.boton_reiniciar]))
        display(self.lista_turnos_output)

    def agregar_turno(self, b):
        nuevo_turno = self.input_turno.value.strip()
        if nuevo_turno:
            self.turnos.append(nuevo_turno)
            self.actualizar_lista_turnos()
            self.input_turno.value = ""  # Limpiar el campo de entrada después de agregar un turno
        else:
            self.mostrar_mensaje("Por favor, ingrese un turno válido.", 'red')

    def mostrar_confirmacion_limpiar(self, b):
        confirmacion_texto = '¿Está seguro que desea limpiar la lista de turnos? (Sí/No)'
        self.mostrar_mensaje(confirmacion_texto, 'black', False)
        input_confirmacion = widgets.Text(value='', description='Respuesta:', style={'description_width': 'initial'})
        display(input_confirmacion)
        boton_confirmar = widgets.Button(description='Confirmar', style={'button_color': 'green'})
        boton_confirmar.on_click(lambda x: self.procesar_confirmacion(input_confirmacion.value.lower()))
        display(boton_confirmar)

    def procesar_confirmacion(self, respuesta):
        if respuesta == 'si':
            self.limpiar_lista()
        elif respuesta == 'no':
            self.ocultar_confirmacion()

    def limpiar_lista(self):
        self.turnos = []
        self.actualizar_lista_turnos()
        self.ocultar_confirmacion()
        self.mostrar_mensaje("La lista de turnos ha sido limpiada.", 'green')
        # Después de limpiar la lista, indicar que el botón "Reiniciar" debe ser visible
        self.boton_reiniciar_visible = True
        self.mostrar_widgets()

    def reiniciar_app(self, b):
        self.turnos = []
        self.actualizar_lista_turnos()
        self.mostrar_mensaje("La aplicación ha sido reiniciada.", 'green')
        # Después de reiniciar, indicar que el botón "Reiniciar" debe ser invisible nuevamente
        self.boton_reiniciar_visible = False
        self.mostrar_widgets()

    def actualizar_lista_turnos(self):
        with self.lista_turnos_output:
            self.lista_turnos_output.clear_output(wait=True)
            for turno in self.turnos:
                print(turno)

    def mostrar_mensaje(self, mensaje, color, limpiar_output=True):
        if limpiar_output:
            clear_output(wait=True)
        mensaje_html = f'<font color="{color}">{mensaje}</font>'
        mensaje_widget = widgets.HTML(value=mensaje_html)
        display(mensaje_widget)

    def ocultar_confirmacion(self):
        clear_output(wait=True)

# Crear la aplicación
app = TurnoApp()
```
